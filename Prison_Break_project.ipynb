We dont' really care about the Details of each escape, so let's remove that column to see the data a little easier.

```python
index = 0
for row in data:
    data[index] = row[:-1]
    index += 1
    
print(data[:3])
    
```

```python
min_year = min(data, key=lambda x: x[0])[0]
max_year = max(data, key=lambda x: x[0])[0]
```

```python
print(max_year)
```

```python
years = []
for y in range(min_year, max_year+1):
    years.append(y)
```

```python
attempts_per_year = []

for y in years:
    attempts_per_year.append([y,0])

print(attempts_per_year)
```

```python
for row in data:
    for ya in attempts_per_year: # Instruction 2 - nothing to do here
        y = ya[0]
        if row[0] == y:
            ya[1] += 1

print(attempts_per_year)
```

```python
barplot(attempts_per_year)
```

```python
sorted_attempts_per_year = sorted(attempts_per_year, reverse = True, key = lambda x: x[1])
print(sorted_attempts_per_year)
```

```python
barplot(sorted_attempts_per_year)
# This does not plot them sorted because of the barplot function
```

## Examine successfulness of attemenpts

```python
print(data[0:2])
```

```python
# column 4 is success and is Yes or No

yes = 0
no = 0

for row in data:
    if row[3] == 'Yes':
        yes += 1
    elif row[3] == 'No':
        no += 1
    else:
        pass
```

```python
print('Overall percent successful: ', round(yes/len(data)*100,2))
```

Now look at at how this rate has changed over the year, though with small number statistics, this won't mean much.

```python
print(years)
```

```python
success_per_year = []

for y in years:
    success_per_year.append([y,0,0])
    
```

```python
for row in data:
    for ya in success_per_year:
        y = ya[0]
        if row[0] == y:
            if row[3] == 'Yes':
                ya[1] += 1
            elif row[3] == 'No':
                ya[2] += 1
            else:
                pass

print(success_per_year)
```

```python
fraction_success_per_year = []

for row in success_per_year:
    if (row[1]+row[2]) == 0: #account for not dividing by zero
        fraction_success_per_year.append([row[0],0])
    else:
        fraction_success_per_year.append([row[0],row[1]/(row[1]+row[2])])
```

```python
print(fraction_success_per_year)
```

```python
barplot(fraction_success_per_year)
```

```python
#print(fraction_success_per_year)

x = []
y = []

for row in fraction_success_per_year:
    x.append(row[0])
    y.append(row[1])
#x = fraction_success_per_year[:][0]
#y = fraction_success_per_year[1]
#print(x)

plt.figure(figsize=(16,9))
sns.barplot(x = x, y = y, color = 'lightcoral')
plt.xticks(rotation=90,fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
