```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2, 2, sharex=True, sharey=True)
axs = [ax1,ax2,ax3,ax4]

ax1.set_xlim(0,10)
ax1.set_ylim(0,0.5)

# generate 4 random variables from the random, gamma, exponential, and uniform distributions
# we will center all of these on 
x1 = np.random.normal(5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)
x4 = np.random.uniform(0,10, 10000)

# Plot the data
ax1.hist(x1, normed=True, bins=20, alpha=0.5,color='firebrick')
ax2.hist(x2, normed=True, bins=20, alpha=0.5,color='firebrick')
ax3.hist(x3, normed=True, bins=20, alpha=0.5,color='firebrick')
ax4.hist(x4, normed=True, bins=20, alpha=0.5,color='firebrick')

#Make it looks nicer
ax1.title.set_text('Normal')
ax2.title.set_text('Gamma')
ax3.title.set_text('Exponential')
ax4.title.set_text('Uniform')
```

```python
# Now, let's animate it

import matplotlib.animation as animation




n = 100
x1 = np.random.normal(5, 1, 2*n)
x2 = np.random.gamma(2, 1.5, 2*n)
x3 = np.random.exponential(2, 2*n)
x4 = np.random.uniform(0,10, 2*n)


# create the function that will do the plotting, where curr is the current frame
def update(curr):
    # check if animation is at the last frame, and if so, stop the animation a
    if curr == n: 
        a.event_source.stop()
    #plt.cla()
    bins = np.arange(0, 10, 1)
    plt.axis([0,10,0,40])
    #plt.annotate('n = {}'.format(curr), [3,27])
    
    
    # Plot the data
    ax1.hist(x1[:curr], bins=bins, alpha=0.5,color='firebrick')
    ax2.hist(x2[:curr], bins=bins, alpha=0.5,color='firebrick')
    ax3.hist(x3[:curr], bins=bins, alpha=0.5,color='firebrick')
    ax4.hist(x4[:curr], bins=bins, alpha=0.5,color='firebrick')

    #Make it looks nicer
    ax1.title.set_text('Normal')
    ax2.title.set_text('Gamma')
    ax3.title.set_text('Exponential')
    ax4.title.set_text('Uniform')
```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2, 2, sharex=True, sharey=True)
axs = [ax1,ax2,ax3,ax4]

ax1.set_xlim(0,10)
ax1.set_ylim(0,40)
#Make it looks nicer
ax1.title.set_text('Normal')
ax2.title.set_text('Gamma')
ax3.title.set_text('Exponential')
ax4.title.set_text('Uniform')

a = animation.FuncAnimation(fig, update)#, interval=1)
```
