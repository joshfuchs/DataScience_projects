```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
```

```python
#Set up array, popular vote count in presidential years
years = np.array([1988,1992,1996,2000,2004,2008,2012,2016,2020])
d_votes = np.array([41809,44909,47401,50999,59028,69498,65915,65853,81268])
r_votes = np.array([48886,39104,39197,50456,62040,59948,60933,62984,74216])
```

# Plot a few Gaussians with slightly different distributions

```python
def gaussian(values,mu, sigma):
    prob = 1. / (sigma * np.sqrt(2. * np.pi)) * np.exp(-0.5 * ((values - mu)/sigma)**2.)
    return prob
```

```python


values = np.linspace(-10,10,num=100)

gaussian1 = gaussian(values,0,2)
gaussian2 = gaussian(values,0,2.1)
gaussian3 = gaussian(values,0,2.2)
gaussian4 = gaussian(values,0,2.3)
gaussian5 = gaussian(values,0,2.4)
gaussian6 = gaussian(values,0,2.5)
gaussian7 = gaussian(values,0,2.6)
gaussian8 = gaussian(values,0,2.7)
gaussian9 = gaussian(values,0,2.8)
gaussian10 = gaussian(values,0,2.9)



```

```python
plt.figure(figsize=(8,6))
plt.plot(values,gaussian1,'b',linewidth=3)
plt.plot(values,gaussian2,'g',linewidth=3)
plt.plot(values,gaussian3,'r',linewidth=3)
plt.plot(values,gaussian4,'c',linewidth=3)
plt.plot(values,gaussian5,'m',linewidth=3)
plt.plot(values,gaussian6,'y',linewidth=3)
plt.plot(values,gaussian7,'k',linewidth=3)
plt.plot(values,gaussian8,'b',linewidth=3)
plt.plot(values,gaussian9,'g',linewidth=3)
plt.plot(values,gaussian10,'r',linewidth=3)
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
```

# Example Bar Chart

```python
plt.figure(figsize=(8,6))

treatmentph_labels = [2,3.5,5.6]
stemlength_labels = [43,46,33]
stemlength_err = [5,6,3]

x_pos = [i for i, _ in enumerate(treatmentph_labels)]
plt.bar(x_pos, stemlength_labels, color='Turquoise', yerr=stemlength_err,error_kw=dict(lw=5, capsize=5, capthick=3))
plt.xlabel("Treatment pH", fontsize=20)
plt.ylabel("Stemlength (mm)",fontsize=20)

plt.xticks(x_pos, treatmentph_labels,fontsize=20)
plt.yticks(fontsize=20)


ax = plt.gca()
ax.axes.spines['right'].set_visible(False)
ax.axes.spines['top'].set_visible(False)
```

# Scatter Plot

```python
scatter_x = np.linspace(1,50,num=50,endpoint=True)
scatter_y = np.sqrt(scatter_x) * np.random.normal(1.0,0.5,50)

plt.figure(figsize=(8,6))

plt.plot(scatter_x,np.abs(scatter_y),'bo',color='blue')
plt.xlabel("Months From Treatment", fontsize=20)
plt.ylabel("Growth Size (cm)",fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

ax = plt.gca()
ax.axes.spines['right'].set_visible(False)
ax.axes.spines['top'].set_visible(False)
```

# Histograms

```python
fig, (ax1,ax2) = plt.subplots(1, 2, sharex=True, sharey=True)
axs = [ax1,ax2]

ax1.set_xlim(0,10)
ax1.set_ylim(0,0.5)

# generate 4 random variables from the random, gamma, exponential, and uniform distributions
# we will center all of these on 
hist_x1 = np.random.normal(5, 1, 10000)
hist_x2 = np.random.gamma(2, 1.5, 10000)


# Plot the data
ax1.hist(hist_x1, density=True, bins=20, alpha=0.5,color='firebrick')
ax2.hist(hist_x2, density=True, bins=20, alpha=0.5,color='firebrick')


#Make it looks nicer
ax1.title.set_text('Normal')
ax2.title.set_text('Gamma')
```

```python
plt.figure(figsize=(8,6))

plt.hist2d(hist_x1,hist_x2,density=True,range=[[0,10],[0,10]],bins=50)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

# Box and Whisker

```python
normal_sample = np.random.normal(loc=0.0, scale=1.0, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)

df = pd.DataFrame({'normal': normal_sample, 
                   'random': random_sample, 
                   'gamma': gamma_sample})


fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(111)

bp = ax.boxplot([ df['normal'], df['random'], df['gamma'] ])#, whis=[0,100])


ax.set_xticklabels(['Normal','Random','Gamma'],fontsize=16,rotation=45)
ax.set_yticklabels(['0','4','8','12','16','20','24','28','32','36'],fontsize=16)


## change outline color, fill color and linewidth of the boxes


plt.setp(bp['boxes'],color='#7570b3',linewidth=2)

plt.setp(bp['fliers'],color='black',marker='o')

plt.setp(bp['medians'],color='#e72955',linewidth=2)

plt.setp(bp['whiskers'],color='#7570b2',linewidth=2)

plt.setp(bp['caps'],color='#7570b2',linewidth=2)


plt.show()
```
