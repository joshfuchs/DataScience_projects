```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
#Set up array, popular vote count in presidential years
years = np.array([1988,1992,1996,2000,2004,2008,2012,2016,2020])
d_votes = np.array([41809,44909,47401,50999,59028,69498,65915,65853,81268])
r_votes = np.array([48886,39104,39197,50456,62040,59948,60933,62984,74216])
```

# 4 plots showing results of presidential elections (labelled)

```python
fig, axs = plt.subplots (2,2,figsize=(16,9))

#top left
labels1 = ['1988','1992','1996','2000','2004','2008','2012','2016','2020']
x1 = np.arange(len(labels1))
width1 = 0.35
rects1 = axs[0,0].bar(x1 - width1/2,d_votes,width1,label='Democratic',color='b')
rects2 = axs[0,0].bar(x1 + width1/2,r_votes,width1,label='Republican',color='r')
axs[0,0].set_xticks(x1)
axs[0,0].set_xticklabels(labels1)
axs[0,0].yaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
axs[0,0].legend(loc='upper left')
axs[0,0].set_title('1,000 votes',loc='left')
axs[0,0].tick_params(labelsize=14)

#top right
axs[0,1].plot(years,d_votes,'b',label='Democratic')
axs[0,1].plot(years,r_votes,'r',label='Republican')
axs[0,1].set_ylim(0,8.5e4)
axs[0,1].set_xticklabels(labels = years, fontsize=14)
axs[0,1].yaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
axs[0,1].legend(loc='lower right')
axs[0,1].set_title('1,000 votes',loc='left')
axs[0,1].tick_params(labelsize=14)

#bottom left
labels = ['2008','2012','2016','2020']
x = np.arange(len(labels))
width = 0.35
rects1 = axs[1,0].bar(x - width/2,d_votes[5:],width,label='Democratic',color='b')
rects2 = axs[1,0].bar(x + width/2,r_votes[5:],width,label='Republican',color='r')
axs[1,0].set_xticks(x)
axs[1,0].set_xticklabels(labels)
axs[1,0].yaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
axs[1,0].legend(loc='upper left')
axs[1,0].set_title('1,000 votes',loc='left')
axs[1,0].tick_params(labelsize=14)


#bottom right
labels3 = ['2008','2012','2016','2020']
x3 = np.arange(len(labels3))
width3 = 0.35
rects1 = axs[1,1].bar(x3 - width3/2,d_votes[5:],width3,label='Democratic',color='b')
rects2 = axs[1,1].bar(x3 + width3/2,r_votes[5:],width3,label='Republican',color='r')
axs[1,1].set_xticks(x3)
axs[1,1].set_xticklabels(labels3)
axs[1,1].set_ylim(58000,82000)
axs[1,1].yaxis.set_major_formatter(ticker.FormatStrFormatter('%d')) #Ensure the y axis is in plain text, not scientific
axs[1,1].legend(loc='upper left')
axs[1,1].set_title('1,000 votes',loc='left')
axs[1,1].tick_params(labelsize=14) #Set axes font size


#axs[1,1].ticklabel_format(style='sci')
#axs[1,1].axis('off') #To turn axis values completely off





```

# Line plots and colors

```python
x_values = np.linspace(0,15,num=15)

y_values1 = x_values*np.random.random_sample(15)
y_values2 = x_values*np.random.random_sample(15)
y_values3 = x_values*np.random.random_sample(15)
y_values4 = x_values*np.random.random_sample(15)


plt.figure(figsize=(8,6))
plt.plot(x_values,y_values1,linewidth=4)
plt.plot(x_values,y_values2,linewidth=4)
plt.plot(x_values,y_values3,linewidth=4)
plt.plot(x_values,y_values4,linewidth=4)

ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)



```

```python
plt.figure(figsize=(8,6))
plt.plot(x_values,y_values1,'r',linewidth=4)
plt.plot(x_values,y_values2,'k',linewidth=4)
plt.plot(x_values,y_values3,'k',linewidth=4)
plt.plot(x_values,y_values4,'k',linewidth=4)

ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
```

# Examples with marker color and symbols

```python
random1 = np.random.random(size=35)
random2 = np.random.random(size=35)

```

```python
plt.figure(figsize=(8,6))
plt.plot(random1,random2,'bo',markersize=20)
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
```

```python
plt.figure(figsize=(8,6))
plt.plot(random1,random2,'bo',markersize=20)
plt.plot(random1[8],random2[8],'ro',markersize=20)
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
```

```python
plt.figure(figsize=(8,6))
plt.plot(random1,random2,'bo',markersize=20)
plt.plot(random1[16],random2[16],'bs',markersize=20)
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
```

```python
plt.figure(figsize=(8,6))
plt.plot(random1[0:17],random2[0:17],'bo',markersize=20)
plt.plot(random1[18:],random2[18:],'rs',markersize=20)
plt.plot(random1[17],random2[17],'ro',markersize=20)
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
```
